```python
import pandas as pd
import re
```

```python
df = pd.read_csv(r'D:\HP\Downloads\job\job_descriptions.csv')
```

```python
df_task3 = df[
    (df['Role'] == 'Data Engineer') &
    (df['Job Title'] == 'Data Scientist') &
    (df['Qualifications'] == 'B.Tech') &
    (df['Job Portal'] == 'LinkedIn') &
    (df['Preference'] == 'Female') &
    (df['latitude'] < 10)
].copy()
```

```python
asia_countries = [
    'Afghanistan', 'Armenia', 'Azerbaijan', 'Bahrain', 'Bangladesh', 'Bhutan', 
    'Brunei', 'Cambodia', 'China', 'Cyprus', 'Georgia', 'India', 'Indonesia', 
    'Iran', 'Iraq', 'Israel', 'Japan', 'Jordan', 'Kazakhstan', 'Kuwait', 
    'Kyrgyzstan', 'Laos', 'Lebanon', 'Malaysia', 'Maldives', 'Mongolia', 
    'Myanmar', 'Nepal', 'North Korea', 'Oman', 'Pakistan', 'Palestine', 
    'Philippines', 'Qatar', 'Saudi Arabia', 'Singapore', 'South Korea', 
    'Sri Lanka', 'Syria', 'Taiwan', 'Tajikistan', 'Thailand', 'Timor-Leste', 
    'Turkey', 'Turkmenistan', 'United Arab Emirates', 'Uzbekistan', 'Vietnam', 'Yemen'
]
```

```python
df_task3 = df_task3[
    (~df_task3['Country'].isin(asia_countries)) & 
    (~df_task3['Country'].str.startswith('C', na=False))
]
```

```python
df_task3['Company Size'] = pd.to_numeric(df_task3['Company Size'], errors='coerce')
df_task3 = df_task3[df_task3['Company Size'] >= 10000]
```

```python
df_task3['Job Posting Date'] = pd.to_datetime(df_task3['Job Posting Date'])
df_task3 = df_task3[
    (df_task3['Job Posting Date'] >= '2023-01-01') & 
    (df_task3['Job Posting Date'] <= '2023-01-06')
]
```

```python
def ends_with_vowel(name):
    return bool(re.search(r'[aeiouAEIOU]$', str(name)))
```

```python
df_task3 = df_task3[df_task3['Contact Person'].apply(ends_with_vowel)]
```

```python
df_task3.to_csv('task3_cleaned.csv', index=False)
print(f"Task 3 filtering complete. Rows found: {len(df_task3)}")
```
