```python
import pandas as pd
import re

```

```python
df = pd.read_csv(r'D:\HP\Downloads\job\job_descriptions.csv')
```

```python
asia_no_i = [
    'Afghanistan', 'Armenia', 'Azerbaijan', 'Bahrain', 'Bangladesh', 'Bhutan', 
    'Brunei', 'Cambodia', 'China', 'Cyprus', 'Georgia', 'Japan', 'Jordan', 
    'Kazakhstan', 'Kuwait', 'Kyrgyzstan', 'Laos', 'Lebanon', 'Malaysia', 
    'Maldives', 'Mongolia', 'Myanmar', 'Nepal', 'North Korea', 'Oman', 
    'Pakistan', 'Palestine', 'Philippines', 'Qatar', 'Saudi Arabia', 'Singapore', 
    'South Korea', 'Sri Lanka', 'Syria', 'Taiwan', 'Tajikistan', 'Thailand', 
    'Timor-Leste', 'Turkey', 'Turkmenistan', 'United Arab Emirates', 'Uzbekistan', 'Vietnam', 'Yemen'
]
```

```python
df_task2 = df[
    (df['Job Title'] == 'Mechanical Engineer') &
    (df['Work Type'].isin(['Full-Time', 'Part-Time'])) &
    (df['Preference'] == 'Male') &
    (df['Job Portal'] == 'Idealist') &
    (df['Country'].isin(asia_no_i))
].copy()
```

```python
df_task2['Sal_Min'] = df_task2['Salary Range'].str.extract(r'\$(\d+)').astype(float) * 1000
df_task2['Exp_Min'] = df_task2['Experience'].str.extract(r'(\d+)').astype(float)
```

```python
df_task2 = df_task2[
    (df_task2['Sal_Min'] > 50000) & 
    (df_task2['Exp_Min'] >= 5) & 
    (df_task2['Company Size'] < 50000)
]
```

```python
def has_min_vowels(name):
    # Counts a, e, i, o, u (case insensitive)
    return len(re.findall(r'[aeiou]', str(name), re.I)) >= 2

df_task2 = df_task2[df_task2['Company'].apply(has_min_vowels)]
```

```python
df_task2.to_csv('task2cleaned.csv', index=False)
print(f"Task 2 complete. {len(df_task2)} rows saved.")
```
