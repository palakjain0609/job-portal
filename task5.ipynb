```python
import pandas as pd

```

```python
# 1. Load Data
df = pd.read_csv(r'D:\HP\Downloads\job\job_descriptions.csv')

# 2. Geography: India and Germany Only
```

```python
df_task5 = df[df['Country'].isin(['India', 'Germany'])].copy()

```

```python
# 3. Categorical Filters
# Qualification: B.Tech | Work Type: Full-Time | Job Portal: Indeed | Preference: Female
# Job Title: Data Scientist, Art Teacher, or Aerospace Engineer
titles = ['Data Scientist', 'Art Teacher', 'Aerospace Engineer']
df_task5 = df_task5[
    (df_task5['Qualifications'] == 'B.Tech') &
    (df_task5['Work Type'] == 'Full-Time') &
    (df_task5['Job Portal'] == 'Indeed') &
    (df_task5['Preference'] == 'Female') &
    (df_task5['Job Title'].isin(titles))
]

#
```

```python
df_task5
```

```python

# Experience > 2 Years | Salary > 10,000 | Date before 08/01/2023
df_task5['Exp_Num'] = df_task5['Experience'].str.extract(r'(\d+)').astype(float)
df_task5['Sal_Min'] = df_task5['Salary Range'].str.extract(r'\$(\d+)').astype(float) * 1000
df_task5['Job Posting Date'] = pd.to_datetime(df_task5['Job Posting Date'])

df_task5 = df_task5[
    (df_task5['Exp_Num'] >= 2) &
    (df_task5['Sal_Min'] > 10000) &
    (df_task5['Job Posting Date'] < '2023-01-08')
]

```

```python
df_task5
```

```python
# 5. Additional Rules: Location not blank & Company Name > 8 characters
df_task5 = df_task5[
    (df_task5['location'].notna()) & 
    (df_task5['location'] != '') &
    (df_task5['Company'].str.len() > 8)
]

```
