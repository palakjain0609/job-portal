```python
import pandas as pd
import re
```

```python

df = pd.read_csv(r'D:\HP\Downloads\job\job_descriptions.csv')

```

```python
df_task6 = df[
    (df['Work Type'] == 'Intern') & 
    (df['latitude'] < 10) & 
    (df['Company Size'] < 50000)
].copy()
```

```python
df_task6 = df_task6[~df_task6['Country'].str.startswith(('A', 'B', 'C', 'D'), na=False)]
```

```python
df_task6 = df_task6[
    (df_task6['Job Title'].str.len() < 10) & 
    (df_task6['Job Title'].str.match(r'^\S+$'))
]
```

```python
df_task6['Exp_Num'] = df_task6['Experience'].str.extract(r'(\d+)').astype(float)
df_task6['Sal_Min'] = df_task6['Salary Range'].str.extract(r'\$(\d+)').astype(float) * 1000
```

```python
df_task6 = df_task6[
    (df_task6['Exp_Num'] % 2 == 0) & 
    (df_task6['Sal_Min'] > 8000)
]
```

```python
df_task6['Job Posting Date'] = pd.to_datetime(df_task6['Job Posting Date'])
df_task6 = df_task6[df_task6['Job Posting Date'].dt.year.isin([2021, 2022, 2023])]
```

```python
df_task6 = df_task6[df_task6['Contact Person'].str.contains('e', case=False, na=False)]
```

```python
df_task6.to_csv('task6_boxplot_data.csv', index=False)
print(f"Task 6 Cleaned: {len(df_task6)} rows found.")
```
