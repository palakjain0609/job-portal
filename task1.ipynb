```python
import pandas as pd
import re
```

```python
df = pd.read_csv(r'D:\HP\Downloads\job\job_descriptions.csv')
```

```python
df_filtered = df[(df['Work Type'] == 'Intern') & (df['latitude'] < 10)].copy()
```

```python
df_filtered = df_filtered[~df_filtered['Country'].str.startswith(('A', 'B', 'C', 'D'), na=False)]
```

```python
df_filtered = df_filtered[
    (df_filtered['Job Title'].str.len() < 10) & 
    (df_filtered['Job Title'].str.match(r'^\S+$'))
    ]
```

```python
df_filtered['Exp_Start'] = df_filtered['Experience'].str.extract(r'(\d+)').astype(float)
df_filtered = df_filtered[df_filtered['Exp_Start'] % 2 == 0]
```

```python
df_filtered['Salary_Min'] = df_filtered['Salary Range'].str.extract(r'\$(\d+)').astype(float) * 1000
df_filtered = df_filtered[df_filtered['Salary_Min'] > 9000]
```

```python
df_filtered['Company Size'] = pd.to_numeric(df_filtered['Company Size'], errors='coerce')
df_filtered = df_filtered[df_filtered['Company Size'] < 50000]
```

```python
df_filtered['Job Posting Date'] = pd.to_datetime(df_filtered['Job Posting Date'])
df_filtered = df_filtered[df_filtered['Job Posting Date'].dt.month % 2 != 0]
```

```python
df_filtered.to_csv('cleaned_data1.csv', index=False)
print(f"Done! Created 'cleaned_data.csv' with {len(df_filtered)} matching rows.")
```
